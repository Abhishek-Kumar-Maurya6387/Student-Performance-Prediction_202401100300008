```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from uploaded CSV file
file_path = '/content/student_data.csv'
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("Dataset Overview:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())


```

```python

# Handling missing values (if any)
print("\nMissing Values:")
print(df.isnull().sum())
```

```python

# Basic statistics of the dataset
print("\nBasic Statistics:")
print(df.describe())
```

```python
# Data Visualization
plt.figure(figsize=(10, 6))
sns.histplot(df[df.columns[1]], bins=20, kde=True, color='blue')  # Assuming first numeric column
plt.title(f'Distribution of {df.columns[1]}')
plt.xlabel(df.columns[1])
plt.ylabel('Frequency')
plt.show()
```

```python
# Data Visualization
plt.figure(figsize=(10, 6))
sns.histplot(df[df.columns[3]], bins=20, kde=True, color='red')  # Assuming first numeric column
plt.title(f'Distribution of {df.columns[3]}')
plt.xlabel(df.columns[3])
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot between two numeric columns (if available)
if len(df.select_dtypes(include=[np.number]).columns) > 1:
    plt.figure(figsize=(8, 5))
    sns.scatterplot(x=df[df.columns[1]], y=df[df.columns[3]], color='black')
    plt.title(f'{df.columns[1]} vs {df.columns[3]}')
    plt.xlabel(df.columns[1])
    plt.ylabel(df.columns[3])
    plt.show()
```

```python
# Display final message
print("Data Analysis Completed!")
```
